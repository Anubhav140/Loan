```python
loan_dataset.drop(['Loan_ID','Dependents'], axis=1, inplace=True)
```

```python
loan_dataset
```

```python
cols=loan_dataset[['Gender', 'Married', 'Self_Employed']]
for i in cols:
  loan_dataset[i].fillna(loan_dataset[i].mode().iloc[0], inplace=True)
```

```python
loan_dataset.isnull().sum()
```

```python
n_cols = loan_dataset[["LoanAmount", "Loan_Amount_Term", "Credit_History"]]
for i in n_cols:
    loan_dataset[i].fillna(loan_dataset[i].mean(axis=0), inplace=True)
```

```python
def bar_chart(col):
    Approved = loan_dataset[loan_dataset["Loan_Status"]=="Y"][col].value_counts()
    Disapproved = loan_dataset[loan_dataset["Loan_Status"]=="N"][col].value_counts()

    df1 = pd.DataFrame([Approved, Disapproved])
    df1.index = ["Approved", "Disapproved"]
    df1.plot(kind="bar")
```

```python
bar_chart("Gender")
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
loan_dataset[["Gender",'Married','Education','Self_Employed','Property_Area','Loan_Status']] = ord_enc.fit_transform(loan_dataset[["Gender",'Married','Education','Self_Employed','Property_Area','Loan_Status']])
loan_dataset.head()
```

```python
loan_dataset[["Gender",'Married','Education','Self_Employed','Property_Area','Loan_Status']] = loan_dataset[["Gender",'Married','Education','Self_Employed','Property_Area','Loan_Status']].astype('int')
```

```python
loan_dataset
```

```python
from sklearn.model_selection import train_test_split
X = loan_dataset.drop("Loan_Status", axis=1)
y = loan_dataset["Loan_Status"]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=2)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

gfc = GaussianNB()
gfc.fit(X_train, y_train)
pred1 = gfc.predict(X_test)
```
